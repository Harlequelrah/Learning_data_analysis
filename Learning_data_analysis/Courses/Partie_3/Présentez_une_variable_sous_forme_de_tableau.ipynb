- X : une variable

- (xi,...,xn) : réalisation observée de X


## *Cas des variables qualitatives et quantitatives discrètes*

X peut prendre plusieurs modalités (k modalités)

## *Cas des variables quantitatives continues*

On agrége les valeurs de X en classe qui seront au nombre de k


## **Représentez les variables sous forme de tableau***

### Pour les variables qualitatives

L'effectif d'une modalité ai est noté ni . La fréquence est le rapport de l'effectif d'une modalité sur l'effectif total.

### Pour les variables quantitatives

#### Variables discrètes
La fréquence cumulée F d'une modalité ai est la somme des fréquences de toutes les modalités inférieurs ou égales à ai .

#### Variables continues

On remplace les modalités par des classes

```python
import numpy as np
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Harlequelrah/Learning_data_analysis/refs/heads/main/Learning_data_analysis/data/operations.csv',parse_dates=[0])
data.head()
```

```python
effectifs = data.value_counts('categ')
effectifs
```

```python
modalites = effectifs.index
modalites
```

```python
tab = pd.DataFrame(modalites,columns=['categ'])
tab
```

```python
tab['n'] = effectifs.values
tab
```

```python
tab['f'] = tab['n']/tab['n'].sum()
tab
```

```python
tab['F'] = tab['f'].cumsum()
tab
```
