# Appréhendez les mesures de forme

## Le skewness empirique

C'est une mesure de d'asymétrie qui correspond à l'étude de régularité ou non avec laquelle les observations se repartissent autour de la valeur centrale . Elle se nomme $\gamma_1$ et son numérateur $ \mu_3 $ . Nous notons `s` l'ecart type.

$\gamma_1 = \frac{\mu_3}{s^3} \ avec \ \mu_3 = \frac{1}{n} \sum_{i=1}^{n}(x_i-\bar{x})^3$


- Si $\gamma_1$ = 0 : alors la distribution est symétrique et $ \mathrm{Med}= \mathrm{Mod} =  \bar{x} $

- Si $\gamma_1$ < 0 : alors la distribution est étalée sur la gauche et $ \mathrm{Mod} > \mathrm{Med} > \bar{x}$


- Si $\gamma_1$ > 0 : alors la distribution est étalée sur la droite et $ \mathrm{Mod} < \mathrm{Med} < \bar{x}$

**Calcul** :

```python
data['montant'].skew()

```

# Le kurtosis empirique

C'est une mesure d'applatissement . Elle s'interprète uniquement pour les distributions symétriques . C'est en fait une comparaison par rapport à la distribution normale (courbe de GAUSS)
Elle est noté $\gamma_2$ et se calcule comme suite :

$\gamma_2 = \frac{\mu_4}{s^4} - 3 $ avec $\mu_4 = \frac{1}{n} \sum_{i=1}^{n} (x_i-\bar{x})^4$

$\mu_3 \ et \  \mu_4$ sont appelés des moments .

**exemple** :

```python
  data['montant'].kurtosis()
```
## Interprétation :    

$\gamma_2$ = 0 : la distribution a le même applatissement que la distribution normale .

$\gamma_2$ < 0 : la distribution est plus applatie que la distribution normale .

$\gamma_2$ > 0 : la distribution est plus moins applatie que la distribution normale .

# Pratique

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Harlequelrah/Learning_data_analysis/refs/heads/main/Learning_data_analysis/data/operations_clean.csv',parse_dates=[0])
data.head()
````

````python
for cat in data['categ'].unique():
  mask = data['categ']==cat
  subset = data[mask]
  print("-"*20)
  print(f"Categorie : {cat}")
  print("Skewness : ",subset['montant'].skew())
  print("Kurtosis : ",subset['montant'].kurtosis())
````
