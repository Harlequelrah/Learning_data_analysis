# Recherchez les corrélations

- La corrélation est une relation entre deux variables . C'est à dire si la connaissance de la valeur de l'une permet de déduire précisement la valeur de l'autre .

- La corrélation n'induit pas la causalité surtout sans experimentation préalable qui respectent certaines conditions.

Paradoxe de simpson : https://www.youtube.com/watch?v=vs_Zzf_vL2I


Si deux évènements A et B sont indépendants alors on s'attend à ce que le nombre d'individus qui satisfont A et B soit égale à $f_a \times g_b$ et different si corrélés .

## Tableau de contingence :

- Chaque valeur du tableau de contingence(hors total) est appelé effectif conjoint

- L'ensemble des effectifs conjoints est appelé distribution conjointe empirique

- les lignes totales sont appelés distribution marginale empirique
