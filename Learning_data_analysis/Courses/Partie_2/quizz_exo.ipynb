```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/Harlequelrah/Learning_data_analysis/refs/heads/main/Learning_data_analysis/data/quizz_exo.csv')
data.head()
```

# Question 1
Sachant que l'ensemble des coureurs proviennent soit de la Seine-Maritime (76), soit de l'Eure (27) ou soit du Calvados (14), quel département n'appartenant pas à cette liste apparaît pourtant dans notre jeu de données ?

```python
data.loc[~data['Dept'].isin([27,76,14])]
# C'est le département 18
```

# Question 2

Notre jeu de données contient plusieurs valeurs manquantes. Déterminez combien exactement et combien de variables sont concernées :

```python
data.isna().sum()
# 2 valeurs manquantes pour le sexe et 4 valeurs manquantes pour l'Age donc 6 variables manquantes sur 2 variables
```

# Question 3

Essayez de convertir la variable Temps dans un format adapté (datetime, par exemple). Vous obtiendrez normalement une valeur manquante, après transformation. À quoi est due cette valeur manquante ?

```python
data['Temps'] = pd.to_datetime(data['Temps'],errors='coerce') # conversion
data['Temps'].isna().sum() # il  y a une variable manquante
# erreur d'irrégularité 54min47s était sa valeur avant la transformation , ses indices (250,2)
```

# Question 4

En regardant les variables quantitatives (Position, Age et Temps_secondes), combien d'outliers se trouvent dans notre jeu de données au total ?

```python
data.loc[data['Position']>340] # on a un oulier pour Position qui a la valeur 2640
```

```python
data.loc[data['Age']>80] # on a un oulier pour Age qui a la valeur 174
```

```python
data['Temps_secondes'] = pd.to_numeric(data['Temps_secondes'],errors='coerce')
data['Temps_secondes'].describe()  # au vu de la sortie on peut suspecter un outlier qui serait le max au vu du std et de la moyenne
```

```python
data.loc[data['Temps_secondes']>4000,'Temps_secondes'] # on a un oulier pour Temps_seconde qui a la valeur 4231.0
# on a au total 3 outliers
```

# Question 5

Notre jeu de données possède-t-il des doublons, soit des personnes ayant la même position et le même temps à l'arrivée, avec les mêmes caractéristiques ?

```python
data.loc[data[['Position','Temps_secondes']].duplicated(keep=False)] # Pas de doublon
```
