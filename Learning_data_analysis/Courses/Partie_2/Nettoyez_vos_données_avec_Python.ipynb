## Traitez les pays



```python
VALID_COUNTRIES= ['France','Cote d\'ivoire','Madagascar','Bénin','Allemagne','USA']
mask= ~ data['pays'].isin(VALID_COUNTRIES)
data.loc[mask,'pays'] = np.nan
```

Traitez les e-mails

- ',' : le séparateur

- n : le nombre de fois où on souhaite séparer

- expand=True : permet de placer les différentes séparations en colonnes

```python
data['email']=data['email'].str.split(',',n=1,expand=True)[0]
data['email']
# comme .str pour chaine de caractère on peut accéder aux méthodes de datetime avec .dt
```

**Traitez les tailles** :

- Convertir les tailles en nombres décimaux.

- Remplacer les valeurs manquantes par la moyenne de notre échantillon

```python
data['taille'] = data['taille'].str[:-1]
data['taille'] = pd.to_numeric(data['taille'],errors='coerce')
data.loc[data['taille'].isnull(),'taille'] = data['taille'].mean()
data['taille']
```

## Traitez les dates

```python
data['date_naissance']=pd.to_datetime(data['date_naissance'],format='%d/%m/%Y',errors='coerce')
```
