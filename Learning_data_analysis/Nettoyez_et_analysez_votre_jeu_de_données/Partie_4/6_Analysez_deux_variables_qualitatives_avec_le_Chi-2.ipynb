# Analysez deux variables qualitatives avec le Chi-2


```python
# représentons d'abord notre analyse par un tableau de contingence
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    "https://raw.githubusercontent.com/Harlequelrah/Learning_data_analysis/refs/heads/main/Learning_data_analysis/Nettoyez_et_analysez_votre_jeu_de_données/data/operations.csv",
    parse_dates=[0],
)
data.head()
data['quart_mois'] = [int((jour-1)*4/31)+1 for jour in data["date_operation"].dt.day]

X = "quart_mois"
Y = "categ"

cont = data.pivot_table(index=X,columns=Y,aggfunc=len,margins=True,margins_name="Total")
cont
```

Ici on recherche une statistique qui nous permette de comparer deux à deux le tableau de contingence et un tableau représentant les $f_i \times n_j$ .

On procédera en faisant la différence qu'on élevera au carré puis qu'on normalisera . On obtient la mesure :
$
 ξ_{i,j} = \frac{(n_{ij}- \frac{n_i \cdot j_i}{n} )^{2} }{\frac{n_i \cdot j_i}{n}}
$

Une représentation où on colore les cases en foncé quand la mesure est grande et en clair quand elle est proche de 0 s'appele carte de chaleur ou heatmap. Si on somme toutes les valeurs de la heatmap on obtient la statistique $\xi_n$ et  $\xi$ se prononce "xi"

$
\xi_{n} = \sum_{i=1}^k \sum_{j=1}^l  \frac{(n_{ij}- \frac{n_i \cdot j_i}{n} )^{2} }{\frac{n_i \cdot j_i}{n}}
$

Normalement on applique à cette mesure un seuil qui nous dira si y a une corrélation mais ce n'est pas toujours évident.
On retiendra que plus "xi" est grand moins l'hypothese d'independance est valide .
Le test avec le seuil  s'appele le test du $\chi^2$ d'independance ("ki-2" en prononciation) et s'ecrit parfois chi-2 ou ki-2 .


Il existe des seuils qu'on peut calculé pour savoir si y a corrélation ou non appelé seuil de significativité .La p-value exprimée en pourcentage permet de nous dire si un test statistique est significatif ou pas.
