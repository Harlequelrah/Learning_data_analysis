# Analysez une variable quantitative et une qualitative par ANOVA

L'ANOVA est une modélisation qui essaie d'expliquer les variations de la variable quantitative en fonction des modalités de la variable qualitative.

Rappel de la formule de la régression linéaire : $Y=aX+b+ϵ$

**exemple sur les opérations bancaires** :
- i : une catégorie d'opération bancaire .
- $\mu$ : montant de reférence des opérations bancaires .
- $\alpha_i$ : ajustement du montant suivant la catégorie de l'opération bancaire .
- $y_i$ : un montant d'un i .

Modèle : $Y=\alpha_i + \mu + \epsilon $

Estimation de $\mu$ noté $\hat{\mu}$ est la moyenne de tous les montants .

Estimation de $\alpha_i$ noté $\hat{\alpha_i} = \bar{y_i} -\hat{\mu}$

**note** : $\alpha_i$ < 0 si en générale les montant d'une categorie < $\mu$ et inversement.


Ce modèle s'appele analyse de la variance :  **ANOVA**

Formule : $SCT = SCE + SCR$

SCT: Variation totale , SCE : variation interclasse et SCR : variation intraclasse

Rapport de corrélation : $η_{X,Y}^{2} = \frac{V_{interclasse}}  {V_{totale}}  $

Si $η_{X,Y}^{2}$ = 0 pas de corrélation à priori mais si 1 y a corrélation

Selon GPT Une grille courante :

- 0 – 0,1 : très faible

- 0,1 – 0,3 : faible à modéré

- 0,3 – 0,5 : modéré

- 0,5 – 0,7 : fort

- `>0,7` : très fort


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    "https://raw.githubusercontent.com/Harlequelrah/Learning_data_analysis/refs/heads/main/Learning_data_analysis/Nettoyez_et_analysez_votre_jeu_de_données/data/operations.csv",
    parse_dates=[0],
)
data.head()
```

# Appréhendez les expressions de SCT, SCE et SCR

ni est l'effectif d'une classe i qui sont au total au nombre de k

---
### SCT : **variation totale ou Total Sum of Squares**
$
SCT = \sum_{i=1}^{k}\sum_{j=1}^{n_i}(y_{ij} - \bar{y})^2
$



---



### SCE : **variation interclasse ou Sum of Squares of the Model**

$
SCE = \sum_{i=1}^{k} n_i ( \bar{y_i} - \bar{y})^2
$


---

### SCR : **variation intraclasse ou Sum of Squares of the Error**
$SCR = \sum_{i=1}^{k} n_i s_i^{2}$ avec $s_i^{2}$ la variance au sein de la classe i

---

```python

X = "categ" # qualitative
Y = "montant" # quantitative

sous_echantillon = data[data["montant"]<0]
sous_echantillon = sous_echantillon.dropna(subset=["categ"])
# Pour récuperer que les dépenses

def eta_squared(x,y):
  moyenne_y = y.mean() # moyenne des montants
  classes = []
  for classe in x.unique(): # boucle sur les catégories
    yi_classe = y[x==classe] # récuperer tous les montants d'une classe
    classes.append({'ni':len(yi_classe),'moyenne_classe':yi_classe.mean()})
  SCT = sum([(yj- moyenne_y)**2 for yj in y])
  SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
  return SCE/SCT

eta_squared(sous_echantillon[X],sous_echantillon[Y])
```
