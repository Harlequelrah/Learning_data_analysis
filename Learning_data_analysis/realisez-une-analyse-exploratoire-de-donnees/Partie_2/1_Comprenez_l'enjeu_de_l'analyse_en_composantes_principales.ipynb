## Objectifs de l'ACP

L'ACP a pour but d'analyser :

- La variabilité entre les individus
- Les liasions entre les variables


## Projection mathématique
C'est représenter des points à p dimensions dans un espace à q dimensions avec q < p .
La projection fait perdre de l'information . L'enjeu est de trouver une projection qui permet de perdre le moins d'information possible .

**note** : des points très étalées ont une grande inertie .

`La clé de l'ACP est de trouver une projection pour laquelle l'inertie des points est maximale.`
