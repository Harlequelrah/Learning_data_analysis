## Objectifs de l'ACP

L'ACP a pour but d'analyser :

- La variabilité entre les individus
- Les liasions entre les variables


## Projection mathématique
C'est représenter des points à p dimensions dans un espace à q dimensions avec q < p .
La projection fait perdre de l'information . L'enjeu est de trouver une projection qui permet de perdre le moins d'information possible .

**note** : des points très étalées ont une grande inertie .

`La clé de l'ACP est de trouver une projection pour laquelle l'inertie des points est maximale.`

## Principe de l'ACP
On recherche les principaux axes d'inertie qui peuvent être vus comme de nouvelles variables.Ils sont aussi la combinaison linéaires des variables initiales.

## Centrage et Reduction
Pour harmoniser dans le cas ou des échantillons ont valeurs de variables dans différentes unités par exemple on fait de sortent que leurs moyennes soient égales et leurs variance aussi.

Classiquement on essaye d'avoir les moyennenes à 0 (centrage) et les variances à 1 (réduction).

### Centrage

On soutire à toutes les observations leurs moyenne.La moyenne des observations centrées vaut alors 0.

### Réduction
Après avoir centré on divise les observations par leurs ecart-type.La variance des observations après réduction vaut alors 1.


Soit la matrice X , $X_{RC}$ la matrice X centrée et réduite , $S_X$ son ecart-type et $\bar{X}$ sa moyenne.

$X_{RC} = \frac{X-\bar{X}}{S_X}$

## Pratique

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
# Permet de standariser les données (centrer et réduire)
```

```python
# Matrice de base :

X = [[12,30,80,-100],
    [-1000,12,-23,10],
    [14,1000,0,0]]

# Version numpy
# X = np.asarrary(X)

# Version pandas:
X = pd.DataFrame(X)
```

```python
X.describe()
```

```python
# Instanciation du scaler :
scaler = StandardScaler()

# On le fit (il apprend les paramètres nécessaires à la transformation):
# scaler.fit(X)

# On l'entraine (transformation des données) :
# X_scaled = scaler.transform(X)

X_scaled = scaler.fit_transform(X)

# On le transforme en dataframe :
X_scaled = pd.DataFrame(X_scaled)

# On peut appliquer .describe() et .round()
X_scaled.describe().round(2)
```
