## Formule du coefficient de corrélation
Soit r le coefficient de corrélation entre deux variables X et Y

$
 r = \frac{ \operatorname{cov}(X,Y)  }{σ_X \cdot σ_Y}
$

# **Coefficient de détermination**

Ce coefficient nous permet de juger de la qualité d'un modèle , c'est à dire à quel point il réussit à expliquer les variations . On dira qu'il est parfait s'il réussi à expliquer la totalité des variations .

Pour le calculer on a besoin des trois mesures suivantes : SCT , SCE , SCR

## **Somme des carrés total (Variation totale T)**
Elle se note SCT .
En anglais anglais SST ( Sum Of Squares Total) .

Soit une variable y .
C'est la somme des carrés des différences entre chaque points ($y_i$) et la moyenne de la variable y ($\bar{y}$) .

### Formule

$SCT = \sum_{i=1}^n(y_i-\bar{y})^2$



## **Somme des carrés expliqués (Variation expliquée par le modèle)**

Elle se note SCE .
En anglais anglais SSR ( Sum Of Squares Regression) .

Soit une variable y .
C'est la somme des carrés des différences entre les points prédits par le modèle ($\hat{y_i}$) et la moyenne des y ($\bar{y}$)

### Formule

$SCE = \sum_{i=1}^n(\hat{y_i}-\bar{y})^2$



## **Somme des carrés résiduels (Variation inexpliquée par le modèle)**

Elle se note SCR .
En anglais anglais SSE ( Sum Of Squares Error) .

Soit une variable y .
C'est la somme des carrés des différences entre les points prédits par le modèle ($\hat{y_i}$) du modèle et les points issus des observations  des y ($y_i$)

### Formule

$SCR = \sum_{i=1}^n(\hat{y_i}-y_i)^2$



On conclura que $SCT = SCE + SCR$  , c'est à dire  variation totale = variation expliquée + variation non expliquée .


# Calcul du coefficient de détermination
Pour la régression linéaire le coefficient de détermination noté $R^{2}$ est en fait le carré du coefficient de corrélation linéaire :
$R^{2} = \frac{SCE}{SCT}$
